```python
# Element-wise multiplication (each element multiplies its equivalent, index 0->0, 1->1, 2->2)
tensor = torch.tensor([1, 2, 3])
print(tensor, "*", tensor)
print("Equals:", tensor * tensor)



# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11],
                         [9, 12]], dtype=torch.float32)

# torch.matmul(tensor_A, tensor_B) # (this will error)



# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T)
print(output)
print(f"\nOutput shape: {output.shape}")
```

```python
# reshape  -   view
import torch
x = torch.arange(1., 8.)
x, x.shape
```

```python
# chnage shape
x_reshaped = x.reshape(1, 7)
x_reshaped, x_reshaped.shape
```

```python
# Change view (keeps same data as original but changes view)
# view requires data to be contigous, if u take transpose then change view will cause error.
z = x.view(1, 7)
z, z.shape
```

```python
# Changing z changes x
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0) # at dim=0 : 4x7 at dim=1: 7x4
x_stacked
```

```python
# VSTACK : Stack tensors in sequence vertically (row wise).
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
print(torch.vstack((a,b)))
a = torch.tensor([[1],[2],[3]])
b = torch.tensor([[4],[5],[6]])
torch.vstack((a,b)) # 3+3 rows of a and b = 6 rows
```

```python
# Hstack: Stack tensors in sequence horizontally (column wise).

a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])
print(torch.hstack((a,b)))
a = torch.tensor([[1],[2],[3]])
b = torch.tensor([[4],[5],[6]])
torch.hstack((a,b))  # 1+1 cols of a and b = 2 cols in result
```

```python
#  Squeeze removes all dims that are 1
print(f"Previous tensor: {x_reshaped}") #x_shaped is 1x7
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")  # just 7
```

```python
#  unsqueeze adds dim1 at specific index
print(f"Previous tensor: {x_squeezed}") #just 7
print(f"Previous shape: {x_squeezed.shape}")

## Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0) # now 1x7  if dim=1: then 7x1
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# using permute to chnages axis
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->2, 1->0, 2->1

print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
# numpy to tensor
import numpy as np
import torch
arr = np.arange(1.,8.)
tensor = torch.from_numpy(arr)  # shares same memory - dtype will be float64 not 32 cuz np has 64 default
arr[0] = 8  # chnage reflected in tensor as this chnages actual memeory
# arr =arr+1 #change doesnt reflect in tensor cuz it creates a new arr its not the old anymore
arr,tensor
```

```python
#tensor to numpy

tensor = torch.ones(7)
np_arr = tensor.numpy()
tensor[0] = 9 # chnage reflected
np_arr,tensor
```

```python
# reproducibility (so you can perform repeatable experiments.)
# when will it gen same nums:
# 1. same pytorch version
# 2. same device (cpu,gpu)
# 3. same seed
randSeed = 0
torch.manual_seed(seed = randSeed)  # each time seed is set to same number it will gen same rand nums
tensor1 = torch.rand(3,4)
torch.manual_seed(seed = randSeed)
tensor2 = torch.rand(3,4)
print(tensor1)
print(tensor2)
print(tensor1 == tensor2)
```
